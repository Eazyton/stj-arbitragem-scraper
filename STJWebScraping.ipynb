```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
import logging
from datetime import datetime

class STJScraperColab:
    def __init__(self):
        self.setup_driver()
        self.setup_logging()
        self.base_url = "https://scon.stj.jus.br/SCON/"

    def setup_driver(self):
        """Configura o driver do Selenium"""
        chrome_options = Options()
        chrome_options.add_argument('--headless')
        chrome_options.add_argument('--no-sandbox')
        chrome_options.add_argument('--disable-dev-shm-usage')
        chrome_options.add_argument('--disable-gpu')
        chrome_options.add_argument('--window-size=1920,1080')

        try:
            self.driver = webdriver.Chrome(options=chrome_options)
            print("Chrome driver inicializado com sucesso!")
        except Exception as e:
            print(f"Erro ao inicializar o Chrome driver: {str(e)}")
            raise

    def setup_logging(self):
        logging.basicConfig(level=logging.INFO)

    def perform_search(self):
        """Realiza a pesquisa no SCON"""
        try:
            print(f"Acessando {self.base_url}...")
            self.driver.get(self.base_url)
            time.sleep(5)

            print("Procurando campo de pesquisa...")
            # Tentar diferentes seletores
            # Substitua a lista de seletores no método perform_search por:
            selectors = [
                'input[placeholder="Digite o(s) critério(s) de pesquisa"]',
                '//input[@placeholder="Digite o(s) critério(s) de pesquisa"]'
            ]

            search_input = None
            for selector in selectors:
                try:
                    if selector.startswith('//'):
                        search_input = WebDriverWait(self.driver, 5).until(
                            EC.presence_of_element_located((By.XPATH, selector))
                        )
                    else:
                        search_input = WebDriverWait(self.driver, 5).until(
                            EC.presence_of_element_located((By.CSS_SELECTOR, selector))
                        )
                    if search_input and search_input.is_displayed() and search_input.is_enabled():
                        print(f"Campo encontrado com seletor: {selector}")
                        break
                except:
                    continue

            if not search_input:
                print("Campo de pesquisa não encontrado!")
                print("HTML da página:", self.driver.page_source[:1000])
                return False

            print('Digitando "ARBITRAGEM"...')
            search_input.clear()
            time.sleep(1)
            search_input.send_keys('"ARBITRAGEM"')
            time.sleep(2)

            print("Enviando pesquisa...")
            search_input.send_keys(Keys.RETURN)
            time.sleep(10)

            print("Aguardando resultados...")
            try:
                WebDriverWait(self.driver, 15).until(
                    EC.presence_of_element_located((By.CLASS_NAME, "documento"))
                )
                print("Resultados encontrados!")
                return True
            except:
                print("Timeout esperando resultados")
                return False

        except Exception as e:
            print(f"Erro durante a pesquisa: {str(e)}")
            print("HTML atual:", self.driver.page_source[:1000])
            return False

    def extract_document_data(self, doc):
      """Extrai dados de um documento"""
      try:
          data = {}

          # Campos básicos que já funcionam
          basic_fields = {
              'processo': ".//div[@class='paragrafoBRS']//div[@class='docTitulo'][contains(text(),'Processo')]/..//div[@class='docTexto']",
              'relator': ".//div[@class='paragrafoBRS']//div[@class='docTitulo'][contains(text(),'Relator')]/..//div[@class='docTexto']",
              'orgao_julgador': ".//div[@class='paragrafoBRS']//div[@class='docTitulo'][contains(text(),'Órgão Julgador')]/..//div[@class='docTexto']",
              'data_julgamento': ".//div[@class='paragrafoBRS']//div[@class='docTitulo'][contains(text(),'Data do Julgamento')]/..//div[@class='docTexto']",
              'data_publicacao': ".//div[@class='paragrafoBRS']//div[@class='docTitulo'][contains(text(),'Data da Publicação')]/..//div[@class='docTexto']",
              'ementa': ".//div[@class='paragrafoBRS']//div[@class='docTitulo'][contains(text(),'Ementa')]/..//div[@class='docTexto']",
              'acordao': ".//div[@class='paragrafoBRS']//div[@class='docTitulo'][contains(text(),'Acórdão')]/..//div[@class='docTexto']",
          }

          # Extrair campos básicos
          for field, xpath in basic_fields.items():
              try:
                  elements = doc.find_elements(By.XPATH, xpath)
                  if elements:
                      data[field] = elements[0].text.strip()
                  else:
                      data[field] = ''
                  print(f"Campo {field}: {data[field][:100]}...")
              except Exception as e:
                  data[field] = ''
                  print(f"Erro ao extrair {field}: {str(e)}")

          # Extrair Referência Legislativa
          try:
              ref_leg_selectors = [
                  "//div[@class='docTitulo'][text()='Referência Legislativa ']/following-sibling::div[@class='docTexto']",
                  "//div[@class='docTitulo'][contains(text(),'Referência')]/following-sibling::div[@class='docTexto']",
                  "//div[contains(@class,'paragrafoBRS')]//pre"
              ]

              refs = []
              for selector in ref_leg_selectors:
                  elements = doc.find_elements(By.XPATH, selector)
                  if elements:
                      # Processar o texto para remover quebras de linha desnecessárias e espaços extras
                      texto = elements[0].text.strip()
                      # Separar por quebras de linha e limpar cada linha
                      linhas = [linha.strip() for linha in texto.split('\n') if linha.strip()]
                      refs.extend(linhas)
                      break

              # Juntar todas as referências com um separador melhor
              data['referencia_legislativa'] = ' | '.join(refs)

          except Exception as e:
              data['referencia_legislativa'] = ''
              print(f"Erro ao extrair referência legislativa: {str(e)}")

          # Extrair Jurisprudência Citada
          try:
              juris_selectors = [
                  "//div[@class='docTitulo'][text()='Jurisprudência Citada ']/following-sibling::div[@class='docTexto']//div[@class='campoVeja']//a",
                  "//div[@class='docTitulo'][contains(text(),'Jurisprudência')]/following-sibling::div[@class='docTexto']//a",
                  "//div[contains(@class,'paragrafoBRS')]//div[@class='campoVeja']//a"
              ]

              jurisprudencias = []
              for selector in juris_selectors:
                  elements = doc.find_elements(By.XPATH, selector)
                  if elements:
                      for element in elements:
                          texto = element.text.strip()
                          # Remover espaços extras e caracteres especiais
                          texto = ' '.join(texto.split())
                          if texto:
                              jurisprudencias.append(texto)
                      if jurisprudencias:
                          break

              # Filtrar duplicatas e juntar com separador
              jurisprudencias = list(dict.fromkeys(jurisprudencias))  # Remover duplicatas
              data['jurisprudencia_citada'] = ' | '.join(jurisprudencias) if jurisprudencias else ''

          except Exception as e:
              data['jurisprudencia_citada'] = ''
              print(f"Erro ao extrair jurisprudência citada: {str(e)}")

          return data

      except Exception as e:
          print(f"Erro ao extrair dados do documento: {str(e)}")
          return None

    def navigate_to_next_page(self):
      """Navega para a próxima página de resultados"""
      try:
          # Procura o botão de próxima página pelo onclick
          next_button = WebDriverWait(self.driver, 9).until(
              EC.element_to_be_clickable((
                  By.CSS_SELECTOR,
                  'a[onclick*="javascript:navegaForm"]'
              ))
          )
          next_button.click()
          time.sleep(9)  # Espera a página carregar
          # Verifica se a página mudou
          print("Navegando para próxima página...")
          return True
      except Exception as e:
          print(f"Não foi possível navegar para a próxima página: {str(e)}")
          return False

    def scrape_acordaos(self, total_documents=467):
      """Realiza o scraping dos documentos"""
      all_acordaos = []

      try:
          if not self.perform_search():
              return pd.DataFrame()

          while len(all_acordaos) < total_documents:
              docs = self.driver.find_elements(By.CLASS_NAME, "documento")
              if not docs:
                  print("Nenhum documento encontrado na página")
                  break

              print(f"\nEncontrados {len(docs)} documentos na página atual")
              current_page = (len(all_acordaos) // 10) + 1
              print(f"Processando página {current_page}")

              for doc in docs:
                  if len(all_acordaos) >= total_documents:
                      break

                  data = self.extract_document_data(doc)
                  if data:
                      all_acordaos.append(data)
                      print(f"Documento {len(all_acordaos)} de {total_documents} extraído com sucesso")

              if len(all_acordaos) < total_documents:
                  print(f"\nMudando para próxima página... ({len(all_acordaos)} documentos extraídos até agora)")
                  if not self.navigate_to_next_page():
                      print("Não foi possível acessar a próxima página")
                      break
                  time.sleep(9)  # Espera maior entre páginas

      except Exception as e:
          print(f"Erro durante o scraping: {str(e)}")

      finally:
          self.driver.quit()
          print("Driver fechado")

      if all_acordaos:
            df = pd.DataFrame(all_acordaos)
            filename = f'acordaos_stj_{datetime.now().strftime("%Y%m%d_%H%M%S")}.csv'
            df.to_csv(filename, index=False, encoding='utf-8-sig')
            print(f"\nScraping concluído! {len(all_acordaos)} acórdãos salvos em {filename}")
            return df
      else:
            print("\nNenhum documento extraído")
            return pd.DataFrame()

def test_scraper(num_docs=467):  # Mudei de 5 para 467
    print(f"Iniciando teste com {num_docs} documentos...")
    try:
        scraper = STJScraperColab()
        df = scraper.scrape_acordaos(total_documents=num_docs)

        if not df.empty:
            print("\nColunas extraídas:")
            print(df.columns.tolist())
            print("\nPrimeiro documento:")
            print(df.iloc[0])

        return df
    except Exception as e:
        print(f"Erro no teste: {str(e)}")
        return None

# Executar teste
if __name__ == "__main__":
    test_scraper(467)  # Aqui também mudei para 467
```

```python
# Primeiro, instalar o xlsxwriter
!pip install xlsxwriter

# Depois o código:
import pandas as pd

def format_csv(input_file, output_file_xlsx):
    """
    Formata o CSV e salva usando xlsxwriter
    """

    # Ler o CSV
    df = pd.read_csv(input_file)

    def cleanup_text(text):
        if pd.isna(text):
            return ''
        text = str(text).replace('\n', ' ')
        text = ' '.join(text.split())
        text = text.replace('*', '').strip()
        return text

    # Limpar todas as colunas
    for column in df.columns:
        df[column] = df[column].apply(cleanup_text)

    # Criar um Excel writer usando xlsxwriter como engine
    writer = pd.ExcelWriter(output_file_xlsx, engine='xlsxwriter')

    try:
        # Converter o DataFrame para Excel
        df.to_excel(writer, index=False, sheet_name='Acordaos')

        # Pegar o workbook e worksheet
        workbook = writer.book
        worksheet = writer.sheets['Acordaos']

        # Ajustar largura das colunas
        for idx, col in enumerate(df.columns):
            series = df[col]
            max_len = max(
                series.astype(str).map(len).max(),  # comprimento máximo no conteúdo
                len(str(col))  # comprimento do cabeçalho
            ) + 2  # adicionar um pouco de espaço
            worksheet.set_column(idx, idx, min(max_len, 100))  # limitar a 100 caracteres

    finally:
        # Garantir que o writer seja fechado
        writer.close()

    print(f"Arquivo Excel salvo em: {output_file_xlsx}")
    print(f"Total de registros processados: {len(df)}")

    return df

# Uso:
df_formatado = format_csv(
    'acordaos_stj_20241101_133025.csv',  # arquivo de entrada
    'acordaos_stj_formatado.xlsx'        # arquivo de saída
)

# Visualizar no Colab
display(df_formatado)
```
